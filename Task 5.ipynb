## Data Visualization

```python
plt.figure(figsize=(12,5))
sns.distplot(df['Sales'],color='b')
plt.xlim(-100,5000)
plt.show()
```

Mostly sales are below 1000

```python
plt.figure(figsize=(12,5))
sns.distplot(df['Quantity'], color= 'b')
plt.show()
```

Mostly customers prefer to purchase 2 to 4 quantity in each transaction

```python
#Analysis of shipping mode
df['Ship Mode'].value_counts()
```

```python
plt.figure(figsize=(11,7))
sns.countplot(x=df['Ship Mode'])
plt.show()
```

```python
sns.pairplot(df, hue="Ship Mode")
```

```python
#Analysis of Segments
df['Segment'].value_counts()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x=df['Segment'])
plt.show()
```

```python
df['City'].value_counts()
```

```python
plt.figure(figsize=(18,9))
sns.countplot(x=df['City'],order=(df['City'].value_counts().head(15)).index)
plt.xticks(rotation=90)
plt.show()
```

```python
df['Category'].value_counts()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x=df['Category'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.bar('Category','Sales',data=df)
plt.title('Sales vs Category',size= 15)
plt.xlabel('Category',size= 13)
plt.ylabel('Sales',size= 13)
```

```python
sns.countplot(x='Category',hue='Region',data=df)
plt.title("Category-Region wise")
```

```python
plt.figure(figsize=[8,6])
ax = sns.barplot(x="Category", y="Profit", data=df, palette="Set2")
```

```python
df['Sub-Category'].value_counts()
```

```python
plt.figure(figsize=(18,9))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.countplot(x='Sub-Category',hue='Region',data=df)
plt.title("Sub-Category-Region wise")
```

```python
plt.figure(figsize=(18,9))
plt.pie(df['Sub-Category'].value_counts(),labels=df['Sub-Category'].value_counts().index, startangle=180, radius=1)
plt.show()
```

```python
plt.figure(figsize=(19,8))
sns.barplot(x=df['Sub-Category'],y=df['Profit'])
plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby(['Segment']).sum()[['Sales','Profit']].plot.bar()
plt.xticks(rotation=45)
plt.title("Sales Vs Profits -Segment wise")
```

```python
df.groupby(['Ship Mode']).sum()[['Sales','Profit']].plot.bar()
plt.xticks(rotation=45)
plt.title("Sales Vs Profits -Ship Mode wise")
```
